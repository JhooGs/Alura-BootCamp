```python
import pandas as pd 
import numpy as np
#Importando biblioteca gráfica do python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
dados = pd.read_csv("./Dados/A160324189_28_143_208.csv", encoding="ISO-8859-1", 
                    skiprows =3, sep=";", skipfooter=12,engine='python',
                    thousands=".",decimal=",")
```

```python
#Formatando globalmente as informações para ter duas casas decimais depois da vírgula
pd.options.display.float_format = "{:.2f}".format
```

```python
dados.plot(x="Unidade da Federação", 
                y="2008/Ago", 
                kind="bar",
                figsize=(12,6))
```

```python
#ax é conhecido como axis
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar",figsize=(12,6))
#Mudar a forma como os dados do eixo Y são exibidos
#x:,.2f => x é só uma letra simbolizando que vai ser alterado, a vírgula é o separador de milhar e .2f é ponto flutuante com duas casas decimais 
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))
plt.title("Valores gastos por unidade da Federação em procedimentos hospitalares")
#Evita que apareça informações que não precisam aparecer
plt.show()
```

```python
#Existe um Dash (-) na informação númerica, transformando o tipo da coluna em object e impossibilitando a plotagem do gráfico
#Aqui eu faço um replace desse caractere para Nan
dados["2020/Ago"].replace('-',np.nan,regex=True, inplace=True)
```

```python
#Transformado a Dash em um valor nulo podemos transformar a coluna em float e plotar
dados["2020/Ago"] =  dados["2020/Ago"].astype(float)
```

```python
dados["2020/Ago"]
```

```python
#ax é conhecido como axis
ax = dados.plot(x="Unidade da Federação", y="2020/Ago", kind="bar",figsize=(12,6))
#Mudar a forma como os dados do eixo Y são exibidos
#x:,.2f => x é só uma letra simbolizando que vai ser alterado, a vírgula é o separador de milhar e .2f é ponto flutuante com duas casas decimais 
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))
plt.title("Valores gastos por unidade da Federação em procedimentos hospitalares")
#Evita que apareça informações que não precisam aparecer
plt.show()
```

```python
#Para chamar mais de uma coluna usamos dois colchetes para simbolizar que são duas nomes distintos 
dados[["2008/Ago","2008/Set"]].head()
```

```python
#Note que a primeira coluna que ele faz a média é 2007, mesmo nosso data frame começando em 1992
#Isso ocorre porque ele só faz o calculo onde todas as colunas possuem todos os dados preenchidos
#Sendo mais específico, a coluna se torna um object caso tenha algo diferente de número, impossibilitando calculos
dados.mean()
```

```python
#Aqui pegamos apenas a primeira coluna e transformamos em uma lista (uma lista permite que a mesma seja alterada depois)
colunas_usaveis = dados.mean().index.tolist()
#Aqui nós vamos alterar a lista e inserir na posição 0 a coluna "Unidade da Federação" (Se a gente não transformar em lista, não poderíamos inserir essa informação)
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```python
#Usando colunas_usaveis como filtro para trazer nosso dataset apenas onde as colunas estão completamente preenchidas
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
#Tranformando o indice da coluna que era 0,1,2,3... nas unidades da federação
usaveis = usaveis.set_index("Unidade da Federação")
```

```python
#vamos dropar a coluna "Total" para deixar o gráfico mais vísivel 
#Axis = 0 É a linha
#Axis = 1 É a coluna
usaveis = usaveis.drop("Total",axis=1)
```

```python
#Gráfico está distorcido por causa do valor do total
usaveis.T.plot(figsize=(14,6))
```

```python
#Desafio 1: Arrumar a legenda

ax = usaveis.T.plot(figsize=(16,6))

plt.title("Valores por Processamento x Data atendimento")
#ncol signigica o número de colunas que a legenda vai ter
#bbox_to_anchor : é a posição no eixo X e Y da legenda, se voc add outra virgula pode colocar largura e altura
leg = ax.legend(loc='lower center', ncol=9, frameon=True,title='Unidades da Federação', bbox_to_anchor=(0.5, -0.35))
#Formatando eixo Y para real
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))
```

```python
#Desafio 2: Plotar 5 estados 

desa2 = usaveis.loc[["41 Paraná","12 Acre","31 Minas Gerais","42 Santa Catarina","53 Distrito Federal"]]
```

```python
ax = desa2.T.plot(figsize=(14,8))
plt.title("Valores por Processamento x Data atendimento")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))
ax.legend(loc='best',frameon=True,title='Unidades da Federação', bbox_to_anchor=(1, 1))
plt.show()
```
