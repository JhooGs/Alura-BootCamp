```python
# index.str.get retorna a string na posição que você estipular
usaveis.index.str.get(3)
```

```python
#Usando regex para localizar campos que comecem com 1, mas o resultado não é muito bom
usaveis.index.str.match('^1')
```

```python
# Vamos criar um dicionário (DePara Região <> Nº)
regiao = {1 : 'Norte',
          2 : 'Nordeste',
          3 : 'Sudeste',
          4 : 'Sul',
          5 : 'Centro-Oeste'
          }
regiao
```

```python
# Cria uma coluna chamada Regiao e atribuímos à ela o número do primeiro elemento do indice
# Exemplo: 12 Acre, primeiro elemento é 1 logo atribuímos 1 a Região

usaveis["Regiao"] = usaveis.index.str.get(0).to_list()
usaveis["Regiao"] = usaveis["Regiao"].astype('int')

usaveis.head()
```

```python
# O map atribui um valor a um elemento
#Como a gente criou um dicionario mostrando o que significa cada número, ele vai se basear nele e substituir o valor da coluna pelo nome

usaveis['Regiao'] = usaveis['Regiao'].map(regiao)
usaveis.head()
```

```python
usaveis.sample(10)
```

```python
#Ordenar pelo Total
ordenado_por_total = usaveis.sort_values("Total", ascending=False)
#Excluindo coluna total para nã odar discrepância na informação
ordenado_por_total = ordenado_por_total.drop("Total", axis = 1)
#Excluir coluna Regiao pois não é numerica para plotar 
ordenado_por_total = ordenado_por_total.drop("Regiao", axis = 1)
ordenado_por_total.head()
```

```python
#DESAFIO 3: DEIXAR MAIS BONITO O GRÁFICO
ax = ordenado_por_total.head(5).T.plot(figsize=(16,8),colormap= 'Dark2', fontsize=14)
plt.title("Estados que gastaram mais")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))
ax.legend(loc='best',frameon=True,title='Unidades da Federação', bbox_to_anchor=(1, 1))

plt.style.use('ggplot')
plt.xticks(rotation=45)
plt.show()
```

```python
#DESAFIO 4: INCLUIR UM ESTADO NO GRÁFICO 

#Estou criando uma variavel e setando nela um slice do data set, como já está ordenado eu só pego os 5 primeiros e transforma em uma lista
ordenado_por_total_mais_estado = ordenado_por_total.T.columns[0:5].tolist()
#Nessa variavel estou colocando mais um estado de minha escolha
ordenado_por_total_mais_estado.append("42 Santa Catarina")
#Aqui eu crio um variavel que recebe o data set inicial com o slice da variavel que eu criei antes, para mostrar os 5 primeiros estados mais o que eu escolhi
meu_estado = ordenado_por_total.loc[ordenado_por_total_mais_estado]
```

```python
ax = meu_estado.T.plot(figsize=(16,8),colormap= 'Dark2', fontsize=14)
plt.title("Estados que gastaram mais")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))
ax.legend(loc='best',frameon=True,title='Unidades da Federação', bbox_to_anchor=(1, 1))

plt.style.use('ggplot')
#rotacionando a legenda
plt.xticks(rotation=45)
plt.show()
```
